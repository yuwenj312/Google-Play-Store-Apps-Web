```python
df = df.drop_duplicates()
```

```python
df
```

```python
#Drop null
values={'Rating':0.0,"Current Ver":'1.0', "Android Ver": '1.0 and up', 'Content Rating': 'Everyone', 'Type':'0'}
df_1 = df.fillna(values)
```

```python
len(sorted(df.App.unique()))
```

```python
# Replace Content Rating column
cr = {'Everyone':0, 'Teen':13, 'Everyone 10+':10, 'Mature 17+':17, 'Adults only 18+':18, 'Unrated':0}
df_1['Content Rating'].replace(cr, inplace=True)
```

```python
# Replace Price column
df_1.Price = df_1['Price'].map(lambda x: float(x.replace('$','')))
```

```python
# Replace installs column
df_1['Installs'].replace({'Free':0}, inplace=True)
df_1.Installs = df_1.Installs.map(lambda x: int(str(x).replace('+', '').replace(',', '')))
```

```python
# Replace Reviews column
df_1.Reviews = df_1.Reviews.map(lambda x: int(x))
```

```python
# Change column name
df_1.rename(columns={"Content Rating":"Content_Rating", "Last Updated": "Last_Updated", "Current Ver": "Current_Ver", "Android Ver": "Android_Ver"}, inplace=True)
```

```python
# Reformat Last_Updated column
df_1['Last_Updated'] = pd.to_datetime(df_1['Last_Updated'],format="%B %d, %Y")
```

```python
df_unique_app = df_1.drop_duplicates(['App'])
df_unique_app['App_id'] = range(1, len(df_unique_app.index)+1)
```

```python
df_unique_app = df_unique_app.reset_index().drop(columns="index")
```

```python
df_unique_app.to_csv("play_store(unique_app).csv", header=True, index=False)
```

```python
df_app = df_1.drop_duplicates(['App', 'Category'])
df_app['App_id'] = range(1, len(df_app.index)+1)
df_app['App_id'] = range(1, len(df_app.index)+1)
df_app = df_app.reset_index().drop(columns="index")
```

```python
df_app.to_csv("play_store.csv", header=True, index=False)
```
